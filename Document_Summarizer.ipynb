```python
# Install required libraries:
# pip install spacy sumy
# python -m spacy download en_core_web_sm

import spacy
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

def summarize_text(text, sentence_count=3):
    # Process text with spaCy
    doc = nlp(text)

    # Convert processed text to plain text for Sumy
    processed_text = " ".join([token.text for token in doc])

    # Use Sumy parser
    parser = PlaintextParser.from_string(processed_text, Tokenizer("english"))

    # LSA Summarizer
    summarizer = LsaSummarizer()

    summary = summarizer(parser.document, sentence_count)

    return " ".join([str(sentence) for sentence in summary])

# Example usage
if __name__ == "__main__":
    text = """
    Artificial Intelligence is transforming the world. It is widely used in healthcare,
    education, finance, and many other industries. AI helps automate tasks and improve
    efficiency. However, there are concerns about job loss and ethical issues.
    Despite challenges, AI continues to evolve rapidly.
    """

    result = summarize_text(text, 2)
    print("Summary:\n", result)
```
